**Movie Rating Prediction**

Movie rating prediction is a type of recommender system that uses machine learning to estimate how much a user would enjoy a movie they haven't seen yet.

How it works:

1.Data Collection:

Movie rating prediction systems rely on a large dataset of movies and their corresponding ratings. This data can come from various sources like online platforms (e.g., IMDB, Netflix), surveys, or even social media.

2.Feature Engineering:
The data is then transformed to include features that might influence ratings. These features can include:

3.Movie characteristics: genre, director, actors, runtime, release year
User characteristics: demographics, watch history, preferred genres
External factors: critical reception, awards
Model Training:
Machine learning algorithms are trained on this data. The goal is to learn the relationship between the features and the ratings. Common algorithms include:

4.Collaborative filtering: Recommends movies similar to what the user liked in the past.
Content-based filtering: Recommends movies with similar features to movies the user liked.
Hybrid approaches: Combine collaborative and content-based methods.
Prediction:
Once the model is trained, it can predict ratings for new movies a user might enjoy.

Uses of Movie Rating Prediction:

1.Recommendation Systems:

Streaming services and online stores use movie rating prediction to suggest movies to users, personalizing their browsing experience.

2.Content Creation and Marketing:
Studios can leverage prediction models to understand audience preferences and tailor their content accordingly.

3.Targeted Advertising:
Platforms can use predictions to recommend movies and target ads to specific user groups.

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

trans_data = movie_data.drop(['Name'],axis=1)
# Transform Director columns
trans_data['Director'] = labelencoder.fit_transform(movie_data['Director'])

# Transform Actors Columns
trans_data['Actor 1'] = labelencoder.fit_transform(movie_data['Actor 1'])
trans_data['Actor 2'] = labelencoder.fit_transform(movie_data['Actor 2'])
trans_data['Actor 3'] = labelencoder.fit_transform(movie_data['Actor 3'])

trans_data['Genre'] = labelencoder.fit_transform(movie_data['Genre'].apply(lambda x: ', '.join(x)))

trans_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
sc_data = scaler.fit_transform(trans_data)
sc_df = pd.DataFrame(sc_data,columns=trans_data.columns)

sc_df.head()
```

```python
# Import modeling Libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
# put data except Rating data
X = trans_data.drop(['Rating'],axis=1)

# Put only Rating data
y = trans_data['Rating']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
X_test = np.array(X_test)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
print('R2 score: ',r2_score(y_test,y_pred))
print('Mean squared error: ',mean_squared_error(y_test,y_pred))
print('Mean absolute error: ',mean_absolute_error(y_test,y_pred))
```
